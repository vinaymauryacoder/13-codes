```python
Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
```

```python
The ArithmeticError class in Python is a base class for all exceptions that occur during arithmetic operations.

Floating Point Error: Raised when a floating-point calculation fails to produce a valid result. This can
    happen when trying to divide a number by zero.
    
# Example of FloatingPointError
try:
    result = 1.0 / 0.0
except FloatingPointError:
    print("Error: division by zero or invalid result")
else:
    print("Result:", result)

    
    
OverflowError: Raised when an arithmetic operation exceeds the maximum representable value for a numeric type. 
    
# Example of OverflowError
try:
    result = 2 ** 1000
except OverflowError:
    print("Error: arithmetic operation resulted in overflow")
else:
    print("Result:", result)
```

```python
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
```

```python
The LookupError class in Python is a base class for all exceptions that occur when trying to access an index or key 
that does not exist in a collection.


KeyError: Raised when trying to access a dictionary key that does not exist.
    
# Example of KeyError
d = {'a': 1, 'b': 2, 'c': 3}

try:
    value = d['d']
except KeyError:
    print("Error: key not found in dictionary")
else:
    print("Value:", value)


IndexError: Raised when trying to access a list or tuple index that does not exist.

# Example of IndexError
l = [1, 2, 3]

try:
    value = l[3]
except IndexError:
    print("Error: index out of range")
else:
    print("Value:", value)
```
